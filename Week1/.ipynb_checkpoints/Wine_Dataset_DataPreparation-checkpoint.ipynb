### Now we can select our target variables. I will select Color Intensity and Flavanoids because the classes have a big enough difference that I can probably get a high accuracy score for when I use the algorithm

```python
groups = data.groupby("Wine class")
for name, group in groups:
    plt.plot(group["Flavanoids"], group["Color intensity"], marker="o", linestyle="", label=name)
plt.legend()
```

### We have defined out x , y axis and  the color of our data. We have scaled it with standart scaler and plotted it into a scatter plot. As we can see class 3 is quite different from the other two but class 2 and 1 are a bit similar, hopefully this will still give us a good result

```python
X_axis = data[['Flavanoids','Color intensity']]
y_axis = data['Wine class']

X_train_iris, X_test_iris, y_train_iris, y_test_iris = train_test_split(X_axis, y_axis,random_state=0)
print('Train shape:', X_train_iris.shape, y_train_iris.shape)
print('Test shape:', X_test_iris.shape, y_test_iris.shape)
```

### We will split the data into train and test so we can use it for out algorithm. So for question 4 we needed to split our data into train and test so we can feed it to the algorithm.  

```python
clf_iris = KNeighborsClassifier(3)
clf_iris.fit(X_train_iris, y_train_iris)

y_pred_iris = clf_iris.predict(X_test_iris)

target_names = ['class 1', 'class 2', 'class 3']
print("Accuracy score per class:",classification_report(y_test_iris, y_pred_iris, target_names=target_names))

print("Accuracy score:" ,metrics.accuracy_score(y_test_iris, y_pred_iris))
```

### We have a fairly high accuracy score which is very nice
